```python
from modules.youtube.extract.ytbpy import channel, playlist, utils, video
from modules import db_manager
from datetime import datetime
from jinja2 import Template
import os

link_channel = 'https://www.youtube.com/@dota2howtosafelane892'

channel_info = channel.get_channel_info(link_channel)

channel_id                  = channel_info["channel_id"].replace("'", "")  # Escape single quotes for SQL
channel_url                 = channel_info["channel_url"].replace("'", "")  # Escape single quotes for SQL
channel_description         = channel_info["description"].replace("'", "")  # Escape single quotes for SQL
channel_title               = channel_info["title"].replace("'", "")  # Escape single quotes for SQL
channel_description_snippet = channel_info["description_snippet"].replace("'", "")  # Escape single quotes for SQL
channel_logo_url            = channel_info["logo_url"].replace("'", "")  # Escape single quotes for SQL
channel_banner_url          = channel_info["banner_url"].replace("'", "") if "banner_url" in channel_info else "" # Escape single quotes for SQL
channel_vanity_url          = channel_info["vanity_url"].replace("'", "")  # Escape single quotes for SQL
channel_handle_name         = channel_info["handle_name"].replace("'", "")  # Escape single quotes for SQL
channel_handle              = channel_info["handle"].replace("'", "")  # Escape single quotes for SQL
channel_game                = 1
channel_country             = None
channel_register            = datetime.now().strftime("%Y-%m-%d")
my_channel                  = False

with open(os.path.join("repository","querys","operation","insert","channel.sql"), "r") as f:
    template = f.read()

sql_template = Template(template)
rendered_sql = sql_template.render(
    channel_id=channel_id,
    channel_url=channel_url,
    channel_description=channel_description,
    channel_title=channel_title,
    channel_description_snippet=channel_description_snippet,
    channel_handle_name=channel_handle_name,
    channel_handle=channel_handle,
    channel_game=channel_game,
    channel_country=channel_country,
    channel_register=channel_register,
    my_channel=my_channel
).replace("'None'", "NULL").replace("'False'", "FALSE").replace("'True'", "TRUE")
try:
    db_manager.execute_raw(rendered_sql)
except Exception as e:
    print(f"Error inserting channel data: {e}")
    db_manager.connection.rollback()
```
